```python
# Mengimpor library yang dibutuhkan
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from IPython.display import Image
```

```python
# Membaca gambar
image = cv2.imread('1398314.jpg')

# Menampilkan gambar asli
print(f"Ukuran file gambar asli: {len(cv2.imencode('.jpg', image)[1]):.0f} bytes ({(len(cv2.imencode('.jpg', image)[1]) / 1024**2):.2f} MB)")
cv2_imshow(image)
```



---



---

Implementasi Kompress Lossy


---



---



```python
# Mengompres gambar menggunakan algoritma lossy dengan OpenCV
encode_param = [int(cv2.IMWRITE_JPEG_QUALITY), 20]  # Nilai kualitas JPEG (0-100), semakin rendah semakin terkompresi
result, encimg = cv2.imencode('.jpg', image, encode_param)
```

```python
# Download gambar terkompresi
from IPython.display import Image
print(f"Ukuran file gambar terkompresi: {len(encimg):.0f} bytes ({(len(encimg) / 1024**2):.2f} MB)")
compressed_image = Image(data=encimg.tobytes(), format='jpeg')
compressed_image
```
